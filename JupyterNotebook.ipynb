Subsetting Data

```python
df = df[['track_name', 'size_bytes', 'currency', 'price', 'rating_count_tot',
       'rating_count_ver', 'user_rating', 'user_rating_ver', 'ver',
       'cont_rating', 'prime_genre']]
```

```python
df.columns
#subsetting columns
```

```python
display(df)
```

Subsetting rows

```python
df_subset1 = df.loc[df['user_rating'] > 3.4]
```

```python
df_subset2 = df.query('user_rating > 3.4')
```

```python
#code for delete null values
df = df.loc[~df['price'].isna()]
```

Casting dtypes

```python
df['rating_count_tot'] = df['rating_count_tot'].astype('float')
```

```python
df['rating_count_tot'] = df['rating_count_tot'].astype('int')
```

```python
df.head()
df['average_rating'] = df['rating_count_ver'] / df['user_rating_ver']
```
